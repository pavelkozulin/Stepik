```python
from sklearn.linear_model import LinearRegression
```

```python
# Для обучения модели используются Х: feature, признак и У: target
# XY есть и у Тренинг, и у Тест
X_train = pd.DataFrame() # Создали пустой датаФрейм
X_train['dayofyear'] = data_train['dayofyear']
X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['dayofyear']

# У остается столбцом (почему????)
y_train = data_train['T']
y_test = data_test['T']
```

```python
plt.scatter(X_train['dayofyear'], y_train, alpha=0.1)
```

```python
# Модель 1: LinearRegression + dayofyear
model = LinearRegression() # Создали пустую модель, параметры a и b - не посчитаны
model.fit(X_train, y_train) # Посчитали a и b
```

```python
pred_train = model.predict(X_train) # Модель использует a и b, чтобы спрогнозировать таргет (на тренировочной выборке)
pred_test = model.predict(X_test) # То же самое для выборки тестовой
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_train, pred_train), mean_squared_error(y_test, pred_test) # Считаем ошибку между Истинным У и спрогнозированным
```

```python
plt.figure(figsize=(25, 10))
plt.plot(data_train['date'], y_train, label='xy_train')
plt.plot(data_train['date'], pred_train, label='xP_train')
plt.plot(data_test['date'], y_test, label='xy_test')
plt.plot(data_test['date'], pred_test, label='xP_test')
plt.legend()
print('Ошибка модели на обучающей выборке', mean_squared_error(y_train, pred_train))
print('Ошибка модели на тестовой выборке', mean_squared_error(y_test, pred_test))
```

```python
import numpy as np
```

```python
data['dayofyear'].min(), data['dayofyear'].max()
```

```python
# LinearRegresion дает плохой прогноз по ПРИЗНАКУ DAYOFYEAR(см График)
# Изначально DAYOFYEAR был корявой подложкой
# Тренировочная выборка конечно больше похожа на поведение СИНУС-КОСИНУС

new_interval = ((data['dayofyear'] - 1) / (365 + 1)) * 2 * np.pi
# Трансформирует ПРИЗНАК в диапазон Тригонометрический (0, 2Пи)

new_interval.min(), new_interval.max()
```

```python
data['cos_dayofyear'] = np.cos(((data['dayofyear'] - 1) / (365+1)) * 2 * np.pi)
# Посчитали КОСИНУС дня в году, чтобы использовать его как ПРИЗНАК для модели
plt.plot(data['date'], data['cos_dayofyear'])
```

```python
# Новая модель !!! 
# LinearRegresion на признаке cos_dayofyear
data['cos_dayofyear'] = np.cos(((data['dayofyear'] - 1) / (365+1)) * 2 * np.pi)

data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

X_train = pd.DataFrame()
X_train['dayofyear'] = data_train['cos_dayofyear'] # Новый признак
X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['cos_dayofyear'] # Новый признак
  
y_train = data_train['T']
y_test = data_test['T']

model = LinearRegression()
model.fit(X_train, y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

plt.figure(figsize=(25, 10))
plt.plot(data_train['date'], y_train, label='xy_train')
plt.plot(data_train['date'], pred_train, label='xP_train')
plt.plot(data_test['date'], y_test, label='xy_test')
plt.plot(data_test['date'], pred_test, label='xP_test')
plt.legend()
print('Ошибка модели на обучающей выборке', mean_squared_error(y_train, pred_train))
print('Ошибка модели на тестовой выборке', mean_squared_error(y_test, pred_test))
```

```python
# ХО-ХО!!!
# Новая модель (гипотеза???)
# Decisin Tree + dayofyear
from sklearn.tree import DecisionTreeRegressor
```

```python
# Повторяем весь PIPE-LINE

data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

X_train = pd.DataFrame()
X_train['dayofyear'] = data_train['dayofyear'] # Старый признак
X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['dayofyear'] # Старый признак

y_train = data_train['T']
y_test = data_test['T']

model = DecisionTreeRegressor() # Новая модель
model.fit(X_train, y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

plt.figure(figsize=(25, 10))
plt.plot(data_train['date'], y_train, label='xy_train')
plt.plot(data_train['date'], pred_train, label='xP_train')
plt.plot(data_test['date'], y_test, label='xy_test')
plt.plot(data_test['date'], pred_test, label='xP_test')
plt.legend()
print('Ошибка модели на обучающей выборке', mean_squared_error(y_train, pred_train))
print('Ошибка модели на тестовой выборке', mean_squared_error(y_test, pred_test))
```

```python
# Ошибка еще упала: Lin + cos было 29-32, Deci + dayofyear: 21-33
# Расхождение между train и test - ПЛОХО, поэтому:
# Модификнём Deci + dayofyear в Deci + dayofyear + настройка параметров ():
# Сейчас Dec имеет много пиков - попробуем ограничить их:

# Повторяем весь предыдущий PIPE-LINE

data_train = data[data['date'] < '2020-01-01']
data_test = data[data['date'] >= '2020-01-01']

X_train = pd.DataFrame()
X_train['dayofyear'] = data_train['dayofyear'] # Старый признак
X_test = pd.DataFrame()
X_test['dayofyear'] = data_test['dayofyear'] # Старый признак

y_train = data_train['T']
y_test = data_test['T']

model = DecisionTreeRegressor(max_depth=6) # Добавляем параметр
model.fit(X_train, y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

plt.figure(figsize=(25, 10))
plt.plot(data_train['date'], y_train, label='xy_train')
plt.plot(data_train['date'], pred_train, label='xP_train')
plt.plot(data_test['date'], y_test, label='xy_test')
plt.plot(data_test['date'], pred_test, label='xP_test')
plt.legend()
print('Ошибка модели на обучающей выборке', mean_squared_error(y_train, pred_train))
print('Ошибка модели на тестовой выборке', mean_squared_error(y_test, pred_test))


```
